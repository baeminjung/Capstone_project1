```python
#https://blog.naver.com/ndb796/221278329032
from sklearn.cluster import KMeans
import numpy as np
import pandas as pd
import seaborn as sb
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.DataFrame(columns =['x', 'y'])
# x위험도 y 수익률
df.loc[0] = [2, 4]
df.loc[1] = [2, 12]
df.loc[2] = [2, 16]
df.loc[3] = [3, 5]
df.loc[4] = [4, 5]
df.loc[5] = [4, 20]
df.loc[6] = [5, 3]
df.loc[7] = [5, 17]
df.loc[8] = [5, 8]
df.loc[9] = [6, 3]
df.loc[10] = [6, 7]
df.loc[11] = [6, 19]
df.loc[12] = [7, 2]
df.loc[13] = [7, 14]
df.loc[14] = [7, 5]
df.loc[15] = [7, 15]
df.loc[16] = [7, 18]
df.loc[17] = [8, 7]
df.loc[18] = [8, 5]
df.loc[19] = [9, 20]
df.loc[20] = [9, 15]
df.loc[21] = [10, 3]
df.loc[22] = [10, 4]
df.loc[23] = [10, 17]
df.loc[24] = [10, 15]
df.loc[25] = [12, 6]
df.loc[26] = [13, 5]
df.loc[27] = [14, 8]
df.loc[28] = [15, 10]
df.loc[29] = [18, 12]

df.head(30)

sb.lmplot('x','y',data=df,fit_reg=False, scatter_kws={"s":100})

plt.title('Fund Recommendation')
plt.xlabel('Risk')
plt.ylabel('Earning Rate')

#클러스터링 시작
points = df.values

kmeans = KMeans(n_clusters=4).fit(points)
kmeans.cluster_centers_
kmeans.labels_

df['cluster'] = kmeans.labels_
df.head(30)

sb.lmplot('x','y',data=df, fit_reg=False, scatter_kws={"s":150}, hue ="cluster")
plt.title('Fund Recommendation')
plt.xlabel('Risk')
plt.ylabel('Earning Rate')
```
