```python
import os
import random
import pandas as pd                        
import numpy as np                        
import fbprophet as Prophet
import datetime
from matplotlib import pyplot
for num in range(9,10):
    df = pd.read_csv('fund_19%d.csv' % num, names=['date', 'rate'], header=0, encoding='cp949')

    #컬럼명 변경
    df.columns = ['ds','y']
    #데이터 타입 변경
    df['ds'] = pd.to_datetime(df['ds'])
    #모델생성
    from fbprophet import Prophet
    model = Prophet()
    #모델 학습
    model.fit(df)
    #list생성
    last_1 = list()
    last_2 = list()
    last_3 = list()

    for i in range(1, 30): #2021 7(31),8(31),9(30)월(3개월간 각각 예측)
        last_1.append(['2021-07-%02d' % i])
    for i in range(1, 31):
        last_2.append(['2021-08-%02d' % i])
    for i in range(1, 31):
        last_3.append(['2021-09-%02d' % i])

    #리스트 형식    
    last_1 = pd.DataFrame(last_1, columns = ['ds'])
    last_1['ds']= pd.to_datetime(last_1['ds'])
    last_2 = pd.DataFrame(last_2, columns = ['ds'])
    last_2['ds']= pd.to_datetime(last_2['ds'])
    last_3 = pd.DataFrame(last_3, columns = ['ds'])
    last_3['ds']= pd.to_datetime(last_3['ds'])


    #예측

    predict1 = model.predict(last_1)
    avg = predict1['yhat']
    #print(sum(avg)/len(avg))
    lower = predict1['yhat_lower']
    #print(sum(lower)/len(lower))
    upper = predict1['yhat_upper']
    #print(sum(upper)/len(upper))

    #print("--------------------")

    predict2 = model.predict(last_2)
    avg1 = predict2['yhat']
    #print(sum(avg)/len(avg))
    lower1 = predict2['yhat_lower']
    #print(sum(lower)/len(lower))
    upper1 = predict2['yhat_upper']
    #print(sum(upper)/len(upper))

    #print("--------------------")

    predict3 = model.predict(last_3)
    avg2 = predict3['yhat']
    #print(sum(avg)/len(avg))
    lower2 = predict3['yhat_lower']
    #print(sum(lower)/len(lower))
    upper2 = predict3['yhat_upper']
    #print(sum(upper)/len(upper))

    print("predict: ", (avg+avg1+avg2)/3)
    print("lower: ", (lower+lower1+lower2)/3)
    print("upper: ", (upper+upper1+upper2)/3)

    #print(predict1[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
    #print(predict2[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
    #print(predict3[['ds', 'yhat', 'yhat_lower', 'yhat_upper']].head())
```
